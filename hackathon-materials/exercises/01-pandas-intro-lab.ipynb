# Lab: Pandas Data Loading & Indexing

Read in the file `data/states.csv`

```python
import pandas as pd
states = pd.read_csv('../data/states.csv')
states
```

Examine the file

```python
states.head()
```

Re-read the file, using the Abbreviation as an index

```python
states.set_index('Abbreviation')
```

Find the largest state by land area

```python
max_area = states.area.max()
states[states.area == max_area]
```

Create a dataframe with just Florida, Georgia, Alabama, and South Carolina

```python
sub_states = states.loc[states.state.isin(['Florida','Georgia','Alabama','South Carolina'])]
sub_states
```

Save the new dataframe to southeast.csv  (there is a `DataFrame.to_csv()` method)

```python
sub_states.to_csv('../data/southeast.csv')
!cat ../data/southeast.csv
```

Read in the sales data `data/sales.csv`

```python
sales = pd.read_csv('../data/sales.csv')
sales.head()
```

Find the average line quantity during the month of June, 2011

```python
june = sales[(sales.date >= '2011-06-01') & (sales.date<'2011-07-01')]
june.qty.mean()
```

Read the stock data `data/closing-prices.csv`

```python
stock = pd.read_csv('../data/closing-prices.csv')
stock.head()
```

```python
stock = pd.read_csv('../data/closing-prices.csv', index_col=0, parse_dates=[0])
```

Examine its structure

```python
dates = stock['Unnamed: 0']
stock.pop('Unnamed: 0')
```

```python
stock['dates'] = dates
stock.set_index('dates')
stock.head()
```

Find the min, max, mean, and median price of AAPL during June of 2015

```python
aapl = stock[['AAPL','dates']]
aapl
june_appl = appl.loc([(appl.dates >= '2015-06-01') & (appl.dates<'2015-07-01')])
june_appl
```
