```python
# Run 'pip install wordcloud' before this cell
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Get text
wcDescriptText = train.Descript.to_string()
wcCategoryText = train.Category.to_string()

# Generate mask
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

# Generate Word Cloud
wc1 = WordCloud(background_color="white", repeat=True, mask=mask).generate(wcDescriptText)
wc2 = WordCloud().generate(wcCategoryText)

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(18, 5))
fig.suptitle('Word Cloud')
ax1.axis("off")
ax2.axis("off")
ax1.title.set_text("Descript")
ax2.title.set_text("Category")
ax1.imshow(wc1, interpolation="bilinear")
ax2.imshow(wc2, interpolation='bilinear')
```
